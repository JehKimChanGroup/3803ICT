```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data_v1.csv", engine="python", on_bad_lines='skip')
print("✅ Loaded rows:", len(df))

# Display the first 5 rows to preview the data
df.head(20)
```

```python
# clean Company for finding additional group
# Drop rows missing Requirement
df_clean = df.dropna(subset=['Requirement'])
```

```python
# create new col CompanyGroupID
df_clean['CompanyGroupID'] = df_clean['Company']


# select only missing row Company
mask_missing_company = df_clean['Company'].isnull()
```

```python
# Mapping from city to state
city_to_state = {
    # New South Wales
    'sydney': 'NSW', 'newcastle': 'NSW', 'wollongong': 'NSW', 'central coast': 'NSW', 'maitland': 'NSW',
    'coffs harbour': 'NSW', 'wagga wagga': 'NSW', 'albury': 'NSW', 'port macquarie': 'NSW',

    # Victoria
    'melbourne': 'VIC', 'geelong': 'VIC', 'ballarat': 'VIC', 'bendigo': 'VIC', 'shepparton': 'VIC',
    'mildura': 'VIC', 'warrnambool': 'VIC', 'traralgon': 'VIC', 'wangaratta': 'VIC',

    # Queensland
    'brisbane': 'QLD', 'gold coast': 'QLD', 'sunshine coast': 'QLD', 'townsville': 'QLD', 'cairns': 'QLD',
    'toowoomba': 'QLD', 'mackay': 'QLD', 'rockhampton': 'QLD', 'bundaberg': 'QLD',

    # South Australia
    'adelaide': 'SA', 'mount gambier': 'SA', 'whyalla': 'SA', 'gawler': 'SA', 'port lincoln': 'SA',
    'murray bridge': 'SA',

    # Western Australia
    'perth': 'WA', 'bunbury': 'WA', 'geraldton': 'WA', 'albany': 'WA', 'kalgoorlie': 'WA', 'broome': 'WA',

    # Tasmania
    'hobart': 'TAS', 'launceston': 'TAS', 'devonport': 'TAS', 'burnie': 'TAS',

    # ACT
    'canberra': 'ACT',

    # Northern Territory
    'darwin': 'NT', 'alice springs': 'NT', 'katherine': 'NT', 'palmerston': 'NT'
}
```

```python
# Helper function to get state from city
def get_state(location):
    location = location.lower()
    for city, state in city_to_state.items():
        if city in location:
            return state
    return 'UnknownState'
```

```python
# Create CompanyGroupID using location/classification
def infer_company(row):
    classification = row['Classification']
    subclassification = row['SubClassification']
    location = row['Location']
    requirement = row['Requirement']

    if pd.isna(classification):
        classification = ''
    if pd.isna(subclassification):
        subclassification = ''
    if pd.isna(location):
        location = ''
    if pd.isna(requirement):
        requirement = ''
    
    # Find state
    state = get_state(location)

    # Retail case
    if 'Retail' in classification or 'Retail' in subclassification:
        return f"Store_{state}"

    # Hospitality case
    if 'Hospitality' in classification or 'Waiting Staff' in subclassification:
        return 'Restaurant'
    
    # Banking case
    if 'Banking' in classification or 'Banking - Retail/Branch' in subclassification:
        return 'Bank'

    # Normal Company case
    if ('Call Centre' in classification or 'Customer Service' in classification or 
        'Call Centre' in subclassification or 'Customer Service' in subclassification):
        return 'Company'
    
    # Others
    if state != 'UnknownState':
        return f"NaN_{state}"
    else:
        return 'NaN'
```

```python
# Apply function to NaN Company rows
df_clean.loc[mask_missing_company, 'CompanyGroupID'] = df_clean[mask_missing_company].apply(infer_company, axis=1)

print("✅ CompanyGroupID created successfully based on State Mapping!")
```
